> 目标：用你自己的城市分区与迁移数据（或一个 toy 数据），构造/估计迁移矩阵 **P**（离散时间）或转移率（连续时间），并分析：

下面的 notebook 提供一个最小模板：你可以替换为真实数据或更复杂的估计方式。


```
import numpy as np

from projects.p02_markov_migration.src.markov import (
    stationary_distribution,
    net_flux,
    detailed_balance_violation,
)

# --- Toy example: build a simple migration transition matrix P (row-stochastic) ---
P = np.array([
    [0.70, 0.20, 0.10],
    [0.15, 0.70, 0.15],
    [0.20, 0.10, 0.70],
], dtype=float)

pi = stationary_distribution(P)
J = net_flux(P, pi=pi)
viol = detailed_balance_violation(P, pi=pi)

pi, viol, J
```
